```python
from PyMultiDictionary import MultiDictionary
from gtts import gTTS
from IPython.display import Audio, display, clear_output
import ipywidgets as widgets

# === Fonction pour faire parler avec gTTS ===
def parler(texte):
    print("üó£Ô∏è Assistant:", texte)
    tts = gTTS(texte, lang='fr')
    tts.save("reponse.mp3")
    display(Audio("reponse.mp3", autoplay=True))

# === Dictionnaire fran√ßais ===
dico = MultiDictionary()

def trouver_synonymes_fr(mot):
    try:
        synonymes = dico.synonym('fr', mot)
        return synonymes if synonymes else []
    except:
        return []

def trouver_contraires_fr(mot):
    try:
        contraires = dico.antonym('fr', mot)
        return contraires if contraires else []
    except:
        return []

# === Widgets (interface) ===
entree_mot = widgets.Text(
    value='',
    placeholder='Entre un mot en fran√ßais (ex: heureux, rapide, gentil)',
    description='Mot:',
    disabled=False,
    layout=widgets.Layout(width='60%')
)

bouton_syno = widgets.Button(description="üîÅ Synonymes", button_style='success')
bouton_contraire = widgets.Button(description="‚ÜîÔ∏è Contraires", button_style='info')
bouton_stop = widgets.Button(description="‚ùå Quitter", button_style='danger')

sortie = widgets.Output()

# === Fonctions des boutons ===
def afficher_synonymes(b):
    with sortie:
        clear_output()
        mot = entree_mot.value.strip().lower()
        if not mot:
            parler("Merci d‚Äôentrer un mot.")
            return
        synonymes = trouver_synonymes_fr(mot)
        if synonymes:
            resultat = ", ".join(synonymes[:5])
            parler(f"Les synonymes de {mot} sont : {resultat}.")
        else:
            parler(f"Je n‚Äôai trouv√© aucun synonyme pour {mot}.")

def afficher_contraires(b):
    with sortie:
        clear_output()
        mot = entree_mot.value.strip().lower()
        if not mot:
            parler("Merci d‚Äôentrer un mot.")
            return
        contraires = trouver_contraires_fr(mot)
        if contraires:
            resultat = ", ".join(contraires[:5])
            parler(f"Les contraires de {mot} sont : {resultat}.")
        else:
            parler(f"Je n‚Äôai trouv√© aucun contraire pour {mot}.")

def quitter(b):
    with sortie:
        clear_output()
        parler("Au revoir et √† bient√¥t !")
        print("üëã Fin du programme.")

# === Liaison des boutons √† leurs fonctions ===
bouton_syno.on_click(afficher_synonymes)
bouton_contraire.on_click(afficher_contraires)
bouton_stop.on_click(quitter)

# === Affichage de l‚Äôinterface ===
parler("Bonjour ! Je suis ton assistant de synonymes et contraires en fran√ßais.")
display(entree_mot, widgets.HBox([bouton_syno, bouton_contraire, bouton_stop]), sortie)
```
